# Creating a New Dataset 
  In this Dataset We see Played total matches by teams and win/loss

```python
team_id_1=[]
win=[]
for i in range(1,(len(ipl_teams)+1)):
    team_id_1.append(len((ipl_matches[(ipl_matches['Team_Name_Id']==i) | (ipl_matches['Opponent_Team_Id']==i)])))
    win.append(len(ipl_matches[ipl_matches['Match_Winner_Id']==i]))
ipl_teams["matches_played"]=team_id_1
ipl_teams['win']=win
ipl_teams['loss']=[team_id_1[i]-win[i] for i in range(13)]
display(ipl_teams)
ipl_teams.to_csv('dataset/Team.csv')
```

# KKR Team Analysis From 2008-2016

```python
team_kkr=ipl_matches[(ipl_matches['Team_Name_Id']==1) | (ipl_matches['Opponent_Team_Id']==1)]
kkr_matches=[]
kkr_loss=[]
kkr={}
for i in range(2,14):
    kkr_matches.append(len(team_kkr[(team_kkr['Team_Name_Id']==i)|(team_kkr['Opponent_Team_Id']==i)]))
    kkr_loss.append(len(team_kkr[team_kkr['Match_Winner_Id']==i]))
kkr['Opponent_Team']=['vs_RCB','vs_CSK','vs_KXIP','vs_RR','vs_DD','vs_MI','vs_DC','vs_KTK','vs_PW','vs_SRH','vs_RPS','vs_GL']
kkr['Matches_Played']=kkr_matches
kkr['Loss']=kkr_loss
import csv
with open('dataset/kkr.csv', 'w') as f:
    for key in kkr.keys():
        f.write("%s,%s\n"%(key,kkr[key]))
kkr=pd.read_csv('dataset/kkr.csv')
display(kkr)
```
